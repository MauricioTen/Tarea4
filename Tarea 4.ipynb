### Diccionarios

```python
#Crea un diccionario que contenga tres claves cuyo valor sea único, una clave
#cuyos valores se encuentren en una lista y una clave cuyos valores se
#encuentren en un diccionario 

MX = {'Nuevo Leon':'Monterey','Jalisco':['Guadalajara','Zapopan','Tlaquepaque'],'Mexico':{'Oaxaca':'Oaxaca','Queretaro':'Queretaro','Puebla':'Cuernavaca'}}
MX
```

```python
#Muestra en pantalla los valores de la clave cuyos valores se encuentran en una lista

print(list(MX.values())[1])
```

```python
#Muestra en pantalla el valor de la segunda clave del diccionario que se encuentra dentro del diccionario principal

Mexico1 = (list(MX.values())[2])  
print(Mexico1)

print(list(Mexico1.values())[1])
```

```python
#Agrega una sexta clave con su valor al diccionario
MX.update({'Morelos':'Cuernavaca'})
print(MX)
```

```python
#Modifica el valor de la tercer clave
#Creamos nuevo dicccionario modificando el valor que queremos actualizar
MX_3rdKey_Updated = {'Mexico': {'Oaxaca': 'Puerto Escondido', 'Queretaro': 'Queretaro', 'Puebla': 'Cuernavaca'}}

#Utilizamos el metdodo update y ponemos la variable con el diccionario actualizado
MX.update(MX_3rdKey_Updated)
print(MX)
```

```python
#Elimina la clave número 2 
MX.pop('Jalisco')
print(MX)
```

```python
#Borra toda la información contenida en el diccionario (No borres el diccionario)
MX.clear()
print(MX)
```
